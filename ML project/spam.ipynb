```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('spam.csv',encoding_errors='ignore')
```

```python
df.head(2)
```

```python
x=df['v2']
y=df['v1']
xtrain,ytrain,xtest,ytest=train_test_split(x,y,train_size=0.5,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
main_pipe=Pipeline(steps=[
    ('pre',TfidfVectorizer()),
    ('model',LogisticRegression())
])
```

```python
main_pipe.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report
```

```python
yred=main_pipe.predict(xtest)
```

```python
print(classification_report(ytest,yred))
```

```python
ypred_xtrain=main_pipe.predict(xtrain)
```

```python
print(classification_report(ypred_xtrain,ytrain))
```
